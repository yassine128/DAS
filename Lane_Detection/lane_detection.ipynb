# Video

```python
import mss
import mss.tools
import numpy as np


def screen_capture(monitor_number=1, scale_factor=0.5):
    laneDet = da.LaneDetection()
    with mss.mss() as sct:
        monitor = sct.monitors[monitor_number]

        while True:
            screenshot = sct.grab(monitor)
            img = np.array(screenshot)

            img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

            # Resize the image
            width = int(img.shape[1] * scale_factor)
            height = int(img.shape[0] * scale_factor)
            dim = (width, height)
            resized_img = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)
            try: 
                laneDet.new_img(resized_img)
                processed_img = laneDet.show_final()
                cv2.imshow('Processed Screen Capture', processed_img)
            except: 
                pass
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cv2.destroyAllWindows()

screen_capture(monitor_number=2)
```
